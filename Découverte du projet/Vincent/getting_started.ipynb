## 2.3 Extract some basic infos of each match into a dataframe row

```python
from time import sleep
m_requests = []
for m_id in last10matchs.json():
    r = requests.get(f'https://europe.api.riotgames.com/lol/match/v5/matches/{m_id}?api_key={RIOT_API_KEY}')
    m_requests.append((m_id,r))
    sleep(0.2)
```

```python
data = []
for m_id,r in m_requests:
    match = r.json()['info']
    team_1 = [p['championName'] for p in match['participants'] if p['teamId'] == 100]
    team_2 = [p['championName'] for p in match['participants'] if p['teamId'] == 200]
    if match['teams'][0]['win'] and (match['teams'][0]['teamId'] == 100):
        winner = 1
    else:
        winner = 2
        
    match_infos = {
        'id': m_id,
        'gameCreation': match['gameCreation'],
        'gameDuration': match['gameDuration'],
        'gameMode': match['gameMode'],
        'team_1': team_1,
        'team_2': team_2,
        'winner': winner
    }
    data.append(match_infos)

pandas.DataFrame(data)
    
    
```

# 3. Pagination and quota management

# 3.1 get all registered matches of the player into a similar dataframe

First get all match ids

```python
all_match_id = []
r = requests.get(f'https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count=100&api_key={RIOT_API_KEY}')
k = 1
errorcount = 0
while len(r.json()) > 0:
    if r.status_code == 200:
        all_match_id += r.json()
        r = requests.get(f'https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start={100*k}&count=100&api_key={RIOT_API_KEY}')
        k+=1
        print(f'Gathered {len(all_match_id)} match ids.')
        errorcount = 0
    elif r.status_code == 429:
        print("Too Many Requests, going to sleep for 2 minutes.")
        sleep(120)
    else:
        print(f'Error : {r.status_code}')
        errorcount += 1
    if errorcount >= 10:
        print("Too many errors. Quitting...")
        break
    sleep(0.1)
print("Ended gathering")
```

```python
from tqdm import tqdm

all_matches_requests = []
i = 0
with tqdm(total=len(all_match_id)) as pbar:
    while i < len(all_match_id):
        m_id = all_match_id[i]
        r = requests.get(f'https://europe.api.riotgames.com/lol/match/v5/matches/{m_id}?api_key={RIOT_API_KEY}')
        if r.status_code == 200:
            all_matches_requests.append((m_id,r))
            r = requests.get(f'https://europe.api.riotgames.com/lol/match/v5/matches/{m_id}?api_key={RIOT_API_KEY}')
            i+=1
            errorcount = 0
            pbar.update(1)
        elif r.status_code == 429:
            print("Too Many Requests, going to sleep for 60 seconds.")
            sleep(60)
        else:
            print(f'Error : {r.status_code}')
            errorcount += 1
        if errorcount >= 10:
            print("Too many errors. Quitting...")
            break
        sleep(0.1)
```
