# 2. Combining multiple api entry points and formating data

## 2.1 Extract the puuid of a player

```python
recup_puuid = f"https://euw1.api.riotgames.com/lol/summoner/v4/summoners/by-name/{PLAYER_NAME}?api_key={RIOT_API_KEY}"
reponse_puuid = requests.get(recup_puuid)
res_puuid = reponse_puuid.json()
print(res_puuid)
puuid = res_puuid["puuid"]
puuid
```

## 2.2 Extract the 10 last matches of the player

```python
nb_games=10
recup_games = f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count={nb_games}&api_key={RIOT_API_KEY}"
reponse_games = requests.get(recup_games)
res_games = reponse_games.json()
res_games
```

## 2.3 Extract some basic infos of each match into a dataframe row

```python

D={}
for game in res_games:
    api_url = f"https://europe.api.riotgames.com/lol/match/v5/matches/{game}?api_key={RIOT_API_KEY}"
    response = requests.get(api_url)
    match = response.json()
    D[game]=match["info"]

df4=pd.DataFrame(D).transpose()
df4
```

# 3. Pagination and quota management

# 3.1 get all registered matches of the player into a similar dataframe

First get all match ids

```python
end = True
count = 0
total_games = []
nb_games = 10
while end:
    recup_games = f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start={count}&count={nb_games}&api_key={RIOT_API_KEY}"
    reponse_games = requests.get(recup_games)
    res_games = reponse_games.json()
    total_games += res_games
    count +=100
    end = (len(res_games)==100)

print(len(total_games))
```

You will need to paginate using proper parameter to get all matches infos. You may go over your quota, check response status code and act properly to get all data. Will take several minutes to execute (more than 6 minute for me), print advancement to speed up debugging.

```python
import time

D={}
for game in total_games:
    api_url = f"https://europe.api.riotgames.com/lol/match/v5/matches/{game}?api_key={RIOT_API_KEY}"
    response = requests.get(api_url)
    while response.status_code==429:
        time.sleep(2)
        api_url = f"https://europe.api.riotgames.com/lol/match/v5/matches/{game}?api_key={RIOT_API_KEY}"
        response = requests.get(api_url)
        print("Waiting for request ...")
    match = response.json()
    D[game]=match["info"]
    
df6 = pd.DataFrame(D).transpose()
df6
```
